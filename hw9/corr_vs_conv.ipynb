```python
import numpy as np
```

```python
def full_corr(k,x):
    pad = len(k)-1
    p = np.hstack([
        np.zeros(pad),
        x,
        np.zeros(pad)]
    )
    return np.array([ np.dot(k, p[i:i+len(k)]) for i in range(len(p)-len(k)+1) ])

def full_conv(a,b):
    return full_corr(np.flip(a), b)
```

```python
k = [1,2]
x = [3,4,5]
print(full_corr(k,x))
print(full_corr(x,k))
print(full_conv(k,x))
print(full_conv(x,k))
```

```python
x = [0,1]
y = [1,2,1]
z = [1,1]
print(full_corr(full_corr(x,y),z))
print(full_corr(x,full_corr(y,z)))

print(full_conv(full_conv(x,y),z))
print(full_conv(x,full_conv(y,z)))
```

```python
x = [2,1,0,4,1,3]
k = [1,3,2,1]
print(full_conv(k,x))
print(full_corr(k,x))
```
