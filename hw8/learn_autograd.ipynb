```python
import autograd.numpy as anp
from autograd import grad, elementwise_grad, jacobian
```

```python
def f11(x):
    return x**2

g = grad(f11)
g(3.0)
```

```python
h = elementwise_grad(f11)
h(anp.array([3.0, 4.0, 5.0]))
```

```python
j = jacobian(f11)
j(anp.array([3.0, 4.0, 5.0]))
```

```python
def f1n(x):
    # [1 ]
    # [2x]
    return anp.array([ x, x**2 ])

g = elementwise_grad(f1n)
g(3.0)
```

```python
j = jacobian(f1n)
j(3.0)
```

```python
j(anp.array([3.0,4.0]))
```

```python
def fnm(X):
    x,y = X
    # [ -1  1 ]
    # [ 2x 2y ]
    return anp.array([ - x + y, x**2 + y**2 ])

X = anp.array([3.0,4.0])
h = elementwise_grad(fnm)
h(X)
```

```python
j = jacobian(fnm)
j(X)
```

```python
def fnm(X):
    x,y = X
    # [ -1  1 ]
    # [ 2x 2y ]
    return anp.array([ lambda x,y: - x + y, lambda x,y:x**2 + y**2 ])

X = anp.array([3.0,4.0])
h = elementwise_grad(fnm)
h(X)
```
